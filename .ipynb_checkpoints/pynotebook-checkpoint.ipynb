```python
import gwinc
import numpy as np
import math
freq = np.logspace(-3, 4, 1000)
budget = gwinc.load_budget('LION', freq)
trace = budget.run()
fig = gwinc.plot_noise(freq, trace)
name="Lion" 
fig.savefig("plots/"+name+".png", bbox_inches='tight')
fields=[freq, trace['Total'][0], trace['QuantumVacuum'][0],trace['Seismic'][0],trace['SuspensionThermal'][0],trace['Coating']['CoatingBrownian'][0],trace['Coating']['CoatingThermoOptic'][0],trace['Substrate']['Total'][0]]
np.savetxt("plots/"+name+".dat", np.transpose(fields), delimiter="\t",header="Freq\tPowerSpec\tQuantum\tSeismic\tSuspensionThermal\tCoatingBrownian\tCoatingThermoOptic")
```
